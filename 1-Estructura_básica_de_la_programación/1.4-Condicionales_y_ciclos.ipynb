### Tortugas y fractales

Podemos combinar ciclos `for`/`while` con ciclos recursivos para dibujar fractales (m谩s precisamente, una aproximaci贸n de ellos) pero al menos una linda aproximaci贸n de ellos) con ayuda de nuestra amiga tortuga.

```julia
function estrella(tort,dist)
    if dist <= 5
        return
    else
        for i in 1:5
            forward(tort,dist)
            estrella(tort,dist/3)
            turn(tort,216)
        end
    end
end

 = Turtle() #Inicializamos nuestra tortuga en el origen del plano.

d = 200

@svg estrella(,d) 
```
